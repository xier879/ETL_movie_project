```python
import requests
from pprint import pprint
from config import api_key
import json
base_URL = "http://www.omdbapi.com/?apikey="

#for i in range (0,1000)
#create a list of IMDB id numbers

responses =[]
IMDB_id=[]

```

```python
#to have a list holding all imdb IDs. 
IMDB_id = new_Bolly_df['imdbId'].tolist()
```

```python

#len(IMDB_id)
#corresponding to daily limits 1000 api calls, set up a range limit
new_IMDB_id = IMDB_id[0:500]
#new_IMDB_id
```

```python
# Establish variables i.e. lists that will become the columns of the df
#movie_id = []
title = []
imdb_rating = []
release_year = []
```

```python
#to check the type object 
#type(responses)
```

```python
#api01-06(bootcamp_classwork)
#to set up a counter indicating the index of responses list 
count=0
#api01-06(bootcamp_classwork)
for id in new_IMDB_id:
    URL= base_URL + api_key + "&i=" + id
    movie_data= requests.get(URL).json()
    responses.append(movie_data)
    #pprint(responses)
    title.append(responses[count]['Title'])
    release_year.append(responses[count]['Year'])
    imdb_rating.append(responses[count]['imdbRating'])
    count+=1

        
```

```python
Bolly_df.columns
```

```python
#to create a dictionary including lists for Ids, titles, ratings, release years
bolly_dic = {
    "imdbId": new_IMDB_id,
    "title": title,
    "Movie_Rating": imdb_rating,
    "releaseYear": release_year

}
bolly_ratings_df = pd.DataFrame(bolly_dic)
```

```python
bolly_ratings_df.head(10)
```

# Join datasets

```python
#join bollywood movie table from Kaggle and new bollywood table(w/ratings)
#activities04-03
# Merge two dataframes using a left join
#left join since there are less rows in bolly_ratints table
merge_bolly_table = pd.merge(bolly_ratings_df, new_Bolly_df, on="imdbId", how="left")
merge_bolly_table.head()
```
